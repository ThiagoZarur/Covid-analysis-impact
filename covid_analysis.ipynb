```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv('/content/drive/MyDrive/coding_platzi/Covid analysis/transformed_data.csv')
data2 = pd.read_csv('/content/drive/MyDrive/coding_platzi/Covid analysis/raw_data.csv')

data['COUNTRY'].value_counts() 
data['COUNTRY'].value_counts().mode() #moda

data2.head()
```

```python
#HDI human development index 

code = data['CODE'].unique().tolist()
country = data['COUNTRY'].unique().tolist()

hdi = []
tc = []
td = []
sti = []
population = data['POP'].unique().tolist()
gdp = []

for i in country:
    hdi.append((data.loc[data['COUNTRY'] == i, 'HDI']).sum()/294)
    tc.append((data2.loc[data2['location'] == i, 'total_cases']).sum())
    td.append((data2.loc[data2['location'] == i, 'total_deaths']).sum())
    sti.append((data.loc[data['COUNTRY'] == i, 'STI']).sum()/294)
    population.append((data2.loc[data2['location'] == i, 'population']).sum()/294)

aggregate_data = pd.DataFrame(list(zip(code, country, hdi, tc, td, sti, population)),
                              columns= ['Country Code', 'Country', 'HDI','Total Cases', 'Total Deaths', 'Stringency Index', 'Population'])
aggregate_data.head()
```

```python
data = aggregate_data.sort_values(by = ['Total Cases'],ascending=False)
highest_covid = data.head(10) 
highest_covid
```

```python
highest_covid["GDP Before Covid"] = [65279.53, 8897.49, 2100.75, 
                            11497.65, 7027.61, 9946.03, 
                            29564.74, 6001.40, 6424.98, 42354.41]

highest_covid["GDP During Covid"] = [63543.58, 6796.84, 1900.71, 
                            10126.72, 6126.87, 8346.70, 
                            27057.16, 5090.72, 5332.77, 40284.64]
highest_covid
```

```python
figure = px.bar(highest_covid, y = 'Total Cases', x = 'Country', title='Countries with Highest Covid Cases')
figure.show()
```

```python
figure = px.bar(highest_covid, y = 'Total Deaths', x = 'Country', title='Countries with Highest Deaths')
figure.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=highest_covid['Country'],
    y=highest_covid['Total Cases'],
    name = 'Total',
    marker_color = 'indianred'
))
fig.add_trace(go.Bar(
    x=highest_covid['Country'],
    y=highest_covid['Total Deaths'],
    name = 'Total Deaths',
    marker_color = 'lightsalmon'
))

fig.update_layout(barmode='group')
fig.show()
```

```python
#Porcentaje de casos con muertes

cases = highest_covid['Total Cases'].sum()
deceased = highest_covid['Total Deaths'].sum()

labels = ['Total Cases', ' Total Deaths']
values = [cases,deceased]

fig = px.pie(highest_covid, values = values, names = labels, title='Percentaje of Total Cases and deaths')
fig.show()
```

```python
fig = px.bar(highest_covid, x='Country', y='Total Cases',
             hover_data=['Population', 'Total Deaths'], 
             color='Stringency Index', height=500, 
             title= "Stringency Index during Covid-19")
fig.show()
```
